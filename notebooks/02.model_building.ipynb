```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("../artifacts/supermarket_inventory_data2.xlsx")
```

```python
sales = data.drop(['DeliveryTimeDays', 'DeliveryStatus'], axis=1)
```

```python
sales.isnull().sum()
```

```python
sales.dropna(subset=['LastSoldDate'], inplace=True)
```

```python
sales.isnull().sum()
```

```python
y = sales['StockQuantity']
y
```

```python
X = sales.drop('StockQuantity', axis=1)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```
